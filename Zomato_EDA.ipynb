## Zomato Dataset Exploratory data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv',encoding = 'latin-1')
df.head(10)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20,10)
sns.heatmap(df.isnull(),yticklabels = False,cbar = False,cmap = 'viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
df=df.merge(df_country, how = 'inner',on = 'Country Code')
df.head(2)
```

```python
df.dtypes
```

```python
df.Country.value_counts()
```

```python
country_names = df.Country.value_counts().index
```

```python
country_val = df.Country.value_counts().values
```

```python
##piechart - top 3 countries
plt.pie(country_val[:3],shadow = True,labels = country_names[:3],colors=['g','b','r'], autopct = "%1.2f%%")
plt.show()
```

##### observation - Zomato has maximum transactions in India followed by US, and UK

```python
df.columns
```

```python
ratings = df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

```python
ratings
```

##### observations -
1. when rating is between 4.5 - 4.9 --> excellent
2. when rating is between 4.0 - 4.4 --> very good
3. when rating is between 3.5 - 3.9 --> good
4. when rating is between 2.5 - 3.4 --> average
5. rating below 2.5 --> poor
6. rating 0 --> not rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20,8)
sns.barplot(x = 'Aggregate rating',  y = 'Rating count', data = ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20,10)
sns.barplot(x = 'Aggregate rating',  y = 'Rating count',hue = 'Rating color', data = ratings, palette = ['black','red','orange','yellow','lightgreen','green'])
```

#### observations
1. 0 rating is highest
2. maximum number of rating asre between 2.5 - 3.5

```python
## count plot - plotted basically wrt categorical variables
sns.countplot(x = "Rating color", data = ratings, palette = ['black','red','orange','yellow','lightgreen','green'] )
```

```python
#find top 10 cuisines
df = df.replace(np.nan,'Dummy')
cus = ','.join(df['Cuisines'].tolist())
cus = cus.replace(" ","").split(',')
pd.value_counts(cus)[:10]
```

```python
#find total vote count for each country and plot a chart for top 10 countries
df.groupby(['Votes','Country']).size().reset_index().sort_values('Votes',ascending = False)
```

```python
votes_c = df.groupby(['Votes','Country']).size().reset_index().sort_values('Votes',ascending = False)
votes_c.groupby(by = 'Country',sort = False).sum()
```

#### observation
maximum votes are there in India

```python
sum_votes = votes_c.groupby(by = 'Country',sort = False).sum()
matplotlib.rcParams['figure.figsize'] = (20,10)
plt.pie(sum_votes.Votes[:10], labels = country_names[:10])
plt.legend()
plt.show()
```
