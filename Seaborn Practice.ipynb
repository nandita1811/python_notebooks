# Seaborn Tutorial

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.corr()
```

```python
print(sns.heatmap(df.corr()))
```

```python
sns.jointplot(x='petal_length',y='sepal_length',data = df,kind='hex',color='red')
```

```python
sns.jointplot(x='petal_length',y='sepal_length',data = df,kind='reg',color='red')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue ='species' )
```

```python
(sns.distplot(df['petal_length']))
```

```python
(sns.distplot(df['petal_length'],kde=False, bins=10))
```

```python
sns.countplot(df['petal_length'])
```

```python
sns.countplot(df['species'])
```

```python
sns.countplot(df['petal_width'])
```

```python
sns.countplot(df['sepal_length'])
```

```python
plt.subplot(2,2,1,frame_on=True)
sns.countplot(df['petal_length'])
plt.subplot(2,2,2)
sns.countplot(y=df['species'])
plt.subplot(2,2,3)
sns.countplot(df['petal_width'])
plt.subplot(2,2,4)
sns.countplot(df['sepal_length'])
```

```python
plt.subplot(2,2,1,frame_on=True)
sns.barplot(x=df['petal_length'],y=df['species'])
plt.subplot(2,2,2)
sns.barplot(x=df['sepal_length'],y=df['species'])
plt.subplot(2,2,3)
sns.barplot(x=df['petal_width'],y=df['species'])
plt.subplot(2,2,4)
sns.barplot(x=df['sepal_length'],y=df['species'])
```

```python
plt.subplot(2,2,1,frame_on=True)
sns.barplot(y=df['petal_length'],x=df['species'])
plt.subplot(2,2,2)
sns.barplot(y=df['sepal_length'],x=df['species'])
plt.subplot(2,2,3)
sns.barplot(y=df['petal_width'],x=df['species'])
plt.subplot(2,2,4)
sns.barplot(y=df['sepal_length'],x=df['species'])
```

```python
plt.subplot(2,2,1,frame_on=True)
sns.boxplot(y=df['petal_length'],x=df['species'])
plt.subplot(2,2,2)
sns.boxplot(y=df['sepal_length'],x=df['species'])
plt.subplot(2,2,3)
sns.boxplot(y=df['petal_width'],x=df['species'])
plt.subplot(2,2,4)
sns.boxplot(y=df['sepal_length'],x=df['species'])
```
