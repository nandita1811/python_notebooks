# Matplotlib tutorial

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,50)
```

```python
##scatter plot
plt.scatter(x,y,c='red')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('grph')
plt.show()
```

```python
plt.plot(x,y)
```

```python
y=x*x
```

```python
plt.plot(x,y)
```

```python
plt.plot(y,x,'ro--')
```

```python
##creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(y,x,'g*--')
```

```python
x=np.arange(0,np.pi*5,0.1)
xsin=np.sin(x)
```

```python
xcos=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,xsin)
plt.subplot(2,1,2)
plt.plot(x,xcos,'r')
```

```python
##barplot
x=[2,8,10]
y=[10,20,60]
plt.bar(x,y)
plt.hist(x,bins=10)
```
